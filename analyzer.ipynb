```python
import matplotlib.pyplot as plt
import barbeque as bq

def analyzer_loop(prob):
    analyzer.import_data()
    analyzer.evaluate_prob(prob)
    analyzer.create_plot()
    analyzer.plot_data()
    analyzer.fit_data()
    analyzer.plot_residuals()
    analyzer.show_plot()
    analyzer.print_plot()
    
def graph_rss_to_qv(rss):
    qv = [8, 8, 8, 16, 16, 32, 32]
    fig = plt.figure()
    plt.title("Backend performances depending on QV", fontsize = 16)
    plt.plot(qv, rss, 'x', label = "quantum backend")
    plt.xticks(qv)
    plt.xlabel("Quantum Volume", fontsize = 12)
    plt.ylabel("Residual Sum of Squares", fontsize = 12)
    plt.show()
    fig.savefig('plots/rss.pdf', bbox_inches='tight')
    
rss = [0.68, 0.599, 0.588, 0.543, 0.324, 0.158, 0.062]
rss_mode_1_c1 = [0.287, 0.782, 0.628, 0.237, 0.658, 0.223, 0.07]
rss_mode_1_c2 = [0.562, 0.789, 0.487, 1.108, 0.984, 0.577, 0.253]
rss_mode_2_c1 = [0.233, 0.471, 0.671, 0.313, 0.423, 0.369, 0.075]
rss_mode_2_c2 = [0.526, 1.24, 0.812, 0.7, 1.284, 0.369, 0.417]
```

```python
graph_rss_to_qv(rss)
```

```python
analyzer = bq.Analyzer(2, [[0,1]], 'qasm_simulator', 10000, 0)
analyzer_loop('00')
print('RSS: %s' % round(analyzer.get_rss(),3))
analyzer = bq.Analyzer(4, [[0,1],[2,3]], 'qasm_simulator', 10000, 1)
analyzer_loop('00')
print('RSS c1: %s' % round(analyzer.get_rss()['c1'],3))
print('RSS c2: %s' % round(analyzer.get_rss()['c2'],3))
analyzer = bq.Analyzer(4, [[0,1],[2,3]], 'qasm_simulator', 10000, 2)
analyzer_loop('00')
print('RSS c1: %s' % round(analyzer.get_rss()['c1'],3))
print('RSS c2: %s' % round(analyzer.get_rss()['c2'],3))
```
