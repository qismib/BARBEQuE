```python
%matplotlib inline

import matplotlib.pyplot as plt
import barbeque as bq
import numpy as np

def analyzerLoop(prob):
    analyzer.importData()
    analyzer.evaluateProb(prob)
    analyzer.createPlot()
    analyzer.plotData()
    analyzer.fitData()
    analyzer.plotResiduals()
    analyzer.showPlot()
    analyzer.printPlot()
    
def graphRSStoQV(rss):
    qv = [8, 8, 8, 16, 16, 32, 32]
    fig = plt.figure()
    plt.title("Backend performances depending on QV", fontsize = 16)
    plt.plot(qv, rss, 'x')
    plt.xticks(qv)
    plt.xlabel("Quantum Volume", fontsize = 12)
    plt.ylabel("Residual Sum of Squares", fontsize = 12)
    plt.show()
    
rss = [0.68, 0.599, 0.588, 0.543, 0.324, 0.158, 0.062]
rssMode1c1 = [0.453, 0.782, 0.628, 0.237, 0.658, 0.223, 0.07]
rssMode1c2 = [21.535, 0.789, 0.487, 1.108, 0.984, 0.577, 0.253]
rssMode2c1 = []
rssMode2c2 = []
```

```python
analyzer = bq.Analyzer(2, [[0,1]], 'qasm_simulator', 10000, 0)
analyzerLoop('00')
print('RSS: %s' % round(analyzer.getRSS(),3))
analyzer = bq.Analyzer(4, [[0,1],[2,3]], 'qasm_simulator', 10000, 1)
analyzerLoop('00')
print('RSS c1: %s' % round(analyzer.getRSS()['c1'],3))
print('RSS c2: %s' % round(analyzer.getRSS()['c2'],3))
analyzer = bq.Analyzer(4, [[0,1],[2,3]], 'qasm_simulator', 10000, 2)
analyzerLoop('00')
print('RSS c1: %s' % round(analyzer.getRSS()['c1'],3))
print('RSS c2: %s' % round(analyzer.getRSS()['c2'],3))
```
